#### Data
This study will utilise data collected from the AT-HYG Database (Augmented Tycho - HYG). 

Now a little dated, the HYG represented the final products of the Hipparcos mission - the Hipparcos and Tycho Catalogues - released in June 1997. The catalogues were constructed under the responsibility of large scientific teams collaborating with ESA (ESA, 1997, The Hipparcos and Tycho Catalogues, ESA SP-1200).The HYG catalogue originated from the Hipparcos mission, a European Space Agency (ESA) project launched in 1989 with the goal of accurately measuring the positions, distances, and proper motions of stars. The mission culminated in the release of the Hipparcos Catalogue in 1997.
The Yale Bright Star Catalogue (5th Edition) was also integrated into the HYG catalogue, providing additional data on star brightness and spectral types.

The more modern AT-HYG Database incorporates the most-up-to-date HYG database (v3.7) and Augmented Tycho-HYG (v2.4) catalogs with the more recently released Gaia DR3 (2020-2022) data from the third data release of the Gaia mission, which provides precise measurements of stellar positions, distances, and motions for over 1.8 billion stars.

The AT-HYG database is an extension of the original Tycho Catalogue released in 2000. Tycho-2 includes improved astrometric and photometric data for over 2.5 million stars. The AT-HYG is invaluable for contemporary astronomical research, offering access to a broader and more accurate dataset covering a wide range of stellar types and distances. Alongside, opportunities to study stellar populations, dynamics, and evolutionary trends with improved precision and completeness compared to earlier catalogues.

Many astronomers frequently utilise the HYG catalogue extensively for various studies, including:
- Understanding stellar populations and distributions within the Milky Way.
- Investigating stellar evolution and dynamics.
- Identifying and characterizing nearby stars and potential candidates for exoplanet searches.

## Feature Engineering - Calculating the key Variables

- Absolute Magnitude Calculation: Since absolute magnitude can be computed using apparent magnitude and distance, we derive this metric as a foundational step.
$ M = m - 5 \cdot (\log_{10}(d)) + 5 $
- Luminosity Estimation: Using the calculated absolute magnitude, we estimate luminosity. This step considers the relationship where luminosity is inversely proportional to absolute magnitude.
$ \frac{L}{L_{\odot}} \approx \left( \frac{M}{M_{\odot}} \right)^{3.5} $
- Radius Estimation: With luminosity known, we estimate the radius of the star. This estimation relies on empirical relationships derived from stellar physics.
$ \frac{R}{R_{\odot}} \approx \left( \frac{M}{M_{\odot}} \right)^{0.8} $
- Temperature Calculation: Finally, temperature is calculated using the estimated luminosity and radius. This step ensures that temperature, a critical variable in our analysis, is derived with a reasonable degree of accuracy despite the initial lack of colour index data.
$ T = \left( \frac{L}{4 \pi R^2 \sigma} \right)^{\frac{1}{4}} $



```python
# Define functions for calculations
def calc_absolute_magnitude(apparent_magnitude, distance):
    return apparent_magnitude - 5 * np.log10(distance) + 5

def calc_luminosity(absolute_magnitude, M_sun=4.83, L_sun=1):
    return L_sun * 10 ** ((M_sun - absolute_magnitude) / 2.5)

def calc_mass(luminosity, L_sun=1, M_sun=1):
    return M_sun * (luminosity / L_sun) ** (1 / 3.5)

def calc_radius(mass, M_sun=1, R_sun=1):
    return R_sun * (mass / M_sun) ** 0.8

def calc_temperature(luminosity, radius, sigma_sb=5.670374419e-8):
    return (luminosity * 3.828e26 / (4 * np.pi * (radius * 6.955e8)**2 * sigma_sb)) ** 0.25
```

```python
# Apply functions to the entire dataset except row 1 
# Row 1 is our sun, and has a distance of 0 and therefore needs to be completed separately 
stars_df.loc[1:, 'AbsoluteMagnitude'] = calc_absolute_magnitude(stars_df.loc[1:, 'ApparentMagnitude'], stars_df.loc[1:, 'Distance(pc)']) # Here I use the slicing operation .loc[1:, ] to indicate to the code that I want to select rows starting from index position 1 (thereby ignore index position 0) through to the end of the data frame.
stars_df.loc[1:, 'Luminosity'] = calc_luminosity(stars_df.loc[1:, 'AbsoluteMagnitude'])
stars_df.loc[1:, 'Mass'] = calc_mass(stars_df.loc[1:, 'Luminosity'])
stars_df.loc[1:, 'Radius'] = calc_radius(stars_df.loc[1:, 'Mass'])
stars_df.loc[1:, 'Temperature'] = calc_temperature(stars_df.loc[1:, 'Luminosity'], stars_df.loc[1:, 'Radius'])

#Inputting values for our sun
# Assign Sun's properties in normalized units
M_sun = 1  # Solar mass unit
R_sun = 1  # Solar radius unit
L_sun = 1  # Solar luminosity unit

# Assign Sun's properties to the first row
stars_df.loc[0, 'AbsoluteMagnitude'] = 4.83 # Now we are using the same .loc[] function but without the slicing operation (:) to indicate to the code that it is only the first row (i.e. row based on the index label 0) which this these values should be edited for 
# These properties will also set the context for the above written functions
stars_df.loc[0, 'Luminosity'] = L_sun  # in units of solar luminosity
stars_df.loc[0, 'Mass'] = M_sun  # in units of solar mass
stars_df.loc[0, 'Radius'] = R_sun  # in units of solar radius

# Calculate Temperature using the function
stars_df.loc[0, 'Temperature'] = calc_temperature(stars_df.loc[0, 'Luminosity'], stars_df.loc[0, 'Radius'])
```

```python
#Create the final key variable, Spectral Class (which is categorical)
def calculate_spectral_class(temperature):
    if temperature >= 30000:
        return 'O'
    elif 10000 <= temperature < 30000:
        return 'B'
    elif 7500 <= temperature < 10000:
        return 'A'
    elif 6000 <= temperature < 7500:
        return 'F'
    elif 5200 <= temperature < 6000:
        return 'G'
    elif 3700 <= temperature < 5200:
        return 'K'
    elif temperature < 3700:
        return 'M'

# Apply the function to create a new column 'SpectralClass'
stars_df['SpectralClass'] = stars_df['Temperature'].apply(calculate_spectral_class) #.apply in this context is a function that enables me to apply my spectral class function along the entire axis of the data frame
```

```python
# Check the normality of key numerical variables
key_variables = ['Luminosity', 'Temperature', 'Mass', 'Radius', 'AbsoluteMagnitude'] # Defining the relevant variables for the correlation analysis 

for variable in key_variables: #For loop: to iterate through the list of key variables and produce the relevant normality statistics for each one
    # Calculate skewness
    skewness = skew(stars_df[variable])
    print(f'Skewness of {variable}: {skewness}')

    # Calculate kurtosis
    kurt = kurtosis(stars_df[variable])
    print(f'Kurtosis of {variable}: {kurt}')

    # Perform Shapiro-Wilk test for normality
    shapiro_test = shapiro(stars_df[variable])
    print(f'Shapiro-Wilk test statistic for {variable}: {shapiro_test.statistic}')
    print(f'Shapiro-Wilk test p-value for {variable}: {shapiro_test.pvalue}')
    print("-" * 150)  # Print line between outputs for better readability
```

The above statistics provide insights in to the shape of the distribution for the key variables Luminosity,Temperature, Mass and Radius

- The Shapiro-Wilk Test (W = 0.0014, p = 5.35e-249) indicate that the Luminosity data significantly deviates from a normal distribution.

- The Shapiro-Wilk Test (W = 0.974, p = 2.15e-121) indicates that the Temperature data deviates from a normal distribution, though less severely than Luminosity.

Mass:

- Kurtosis: The high kurtosis value ($\gamma_2$ = 37.286) suggests the presence of significant outliers, leading to a distribution with heavy tails.

The current sample size (2,491,420) is substantially above this threshold, and therefore, further visual inspection of these variables is required to assess their normality.

### Box-Cox Transformations for Normality

```python
# Apply Box-Cox transformation to Luminosity
transformed_luminosity, lambda_luminosity = boxcox(stars_df['Luminosity'])

# Calculate Skewness and Kurtosis of transformed data
transformed_lum_skewness = skew(transformed_luminosity)
transformed_lum_kurtosis = kurtosis(transformed_luminosity)
print(f'Transformed Skewness of Luminosity: {transformed_lum_skewness}')
print(f'Transformed Kurtosis of Luminosity: {transformed_lum_kurtosis}')

# Shapiro-Wilk test for normality of transformed data
shapiro_test_transformed = shapiro(transformed_luminosity)
print(f'Shapiro-Wilk test statistic for transformed Luminosity: {shapiro_test_transformed.statistic}')
print(f'Shapiro-Wilk test p-value for transformed Luminosity: {shapiro_test_transformed.pvalue}')
print("-" * 150)  # Print line between outputs for better readability

# Apply Box-Cox transformation to Temperature
transformed_temperature, lambda_temperature = boxcox(stars_df['Temperature'])

# Calculate Skewness and Kurtosis of transformed data
transformed_temp_skewness = skew(transformed_temperature)
transformed_temp_kurtosis = kurtosis(transformed_temperature)
print(f'Transformed Skewness of Temperature: {transformed_temp_skewness}')
print(f'Transformed Kurtosis of Temperature: {transformed_temp_kurtosis}')

# Shapiro-Wilk test for normality of transformed data
shapiro_test_transformed = shapiro(transformed_temperature)
print(f'Shapiro-Wilk test statistic for transformed Temperature: {shapiro_test_transformed.statistic}')
print(f'Shapiro-Wilk test p-value for transformed Temperature: {shapiro_test_transformed.pvalue}')
print("-" * 150)  # Print line between outputs for better readability

# Apply Box-Cox transformation to Radius
transformed_radius, lambda_radius = boxcox(stars_df['Radius'])

# Calculate Skewness and Kurtosis of transformed data
transformed_radius_skewness = skew(transformed_radius)
transformed_radius_kurtosis = kurtosis(transformed_radius)
print(f'Transformed Skewness of Radius: {transformed_radius_skewness}')
print(f'Transformed Kurtosis of Radius: {transformed_radius_kurtosis}')

# Shapiro-Wilk test for normality of transformed data
shapiro_test_transformed = shapiro(transformed_radius)
print(f'Shapiro-Wilk test statistic for transformed Radius: {shapiro_test_transformed.statistic}')
print(f'Shapiro-Wilk test p-value for transformed Radius: {shapiro_test_transformed.pvalue}')
print("-" * 150)  # Print line between outputs for better readability

# Apply Box-Cox transformation to Mass
transformed_mass, lambda_mass = boxcox(stars_df['Mass'])

# Calculate Skewness and Kurtosis of transformed data
transformed_mass_skewness = skew(transformed_mass)
transformed_mass_kurtosis = kurtosis(transformed_mass)
print(f'Transformed Skewness of Mass: {transformed_mass_skewness}')
print(f'Transformed Kurtosis of Mass: {transformed_mass_kurtosis}')

# Shapiro-Wilk test for normality of transformed data
shapiro_test_transformed = shapiro(transformed_mass)
print(f'Shapiro-Wilk test statistic for transformed Mass: {shapiro_test_transformed.statistic}')
print(f'Shapiro-Wilk test p-value for transformed Mass: {shapiro_test_transformed.pvalue}')
print("-" * 150)  # Print line between outputs for better readability

# Adjust AbsoluteMagnitude data to ensure it's positive
adjusted_absolute_magnitude = stars_df['AbsoluteMagnitude'] - stars_df['AbsoluteMagnitude'].min() + 1e-6 # The boxcox transformation requires that the data be strictly positive. Absolute magnitude values can sometimes include negative values for brighter objects and therefore adjusted_absolute_magnitude shifts the AbsoluteMagnitude values by subtracting the minimum value and adding a small constant (1e-6) to ensure all values are positive.

# Apply Box-Cox transformation to adjusted AbsoluteMagnitude
transformed_absolute_magnitude, lambda_absolute_magnitude = boxcox(adjusted_absolute_magnitude)

# Calculate Skewness and Kurtosis of transformed data
transformed_absolute_magnitude_skewness = skew(transformed_absolute_magnitude)
transformed_absolute_magnitude_kurtosis = kurtosis(transformed_absolute_magnitude)
print(f'Transformed Skewness of AbsoluteMagnitude: {transformed_absolute_magnitude_skewness}')
print(f'Transformed Kurtosis of AbsoluteMagnitude: {transformed_absolute_magnitude_kurtosis}')

# Shapiro-Wilk test for normality of transformed data
shapiro_test_transformed = shapiro(transformed_absolute_magnitude)
print(f'Shapiro-Wilk test statistic for transformed AbsoluteMagnitude: {shapiro_test_transformed.statistic}')
print(f'Shapiro-Wilk test p-value for transformed AbsoluteMagnitude: {shapiro_test_transformed.pvalue}')
print("-" * 150)  # Print line between outputs for better readability
```

- These new values suggest that the Box-Cox transformations have normalized the skewness and kurtosis values significantly close to zero, indicating a much more symmetric and less peaked distribution compared to the original data. Additionally, the Shapiro-Wilk test results for each transformed variable show very small p-values, indicating strong evidence against the null hypothesis of normality. 
- This suggests that even after transformation, the data deviates significantly from a normal distribution, likely due to the large sample size, where small deviations from normality can still yield very low p-values.
- These results suggest that while the Box-Cox transformations have helped in reducing skewness and kurtosis, the data still might not be normally distributed, which is important to consider when deciding upon further analytical techniques.

Positive values indicate positions north of the equator, while negative values indicate southward positions. The celestial equator sits at 0 degrees Dec, with the North Celestial Pole at +90 degrees Dec and the South Celestial Pole at -90 degrees Dec. Together, RA and Dec pinpoint the exact locations of celestial objects.

The sample data on the position and apparent magnitude of stars was acquired from the AT-HYG database. After handling the missing data, the final sample size was 2,491,418 stars.

Key variables such as absolute magnitude, luminosity, mass, radius, and temperature were calculated across the database. Descriptive statistics for these variables revealed significant differences between the mean ($\mu$) and median values ($\tilde{M}$), indicating potential skewness. For instance, the temperature had a mean of 8032.93, a standard deviation ($\sigma$) of 1758.67, and a median of 7827.61, while luminosity had a mean of 41.45, a standard deviation of 2384.28, and a median of 9.43. Similar patterns were observed for mass ($\mu$ = 2.12, $\sigma$ = 1.02, $\tilde{M}$ = 1.90), radius ($\mu$ = 1.79, $\sigma$ = 0.68, $\tilde{M}$ = 1.67), and absolute magnitude ($\mu$ = 2.38, $\sigma$ = 2.39, $\tilde{M}$ = 2.39).

Skewness ($\gamma_1$), kurtosis ($\gamma_2$), and Shapiro-Wilk ($ W $) tests were conducted to assess normality, and the results confirmed significantly non-normal distributions for all variables. For example, temperature had a skewness of 2.11, kurtosis of 8.54, and a Shapiro-Wilk statistic of 0.82 (p < 0.001). Visual inspection using histograms, Q-Q plots, and box plots further supported these findings.

Box-Cox transformations were attempted to reduce skewness and kurtosis. Although these transformations significantly reduced both skewness and kurtosis, all variables remained significantly non-normally distributed. For example, the transformed temperature had a mean of 16.36, a standard deviation of 0.65, and a median of 16.34, while the transformed luminosity had a mean of 2.33, a standard deviation of 1.66, and a median of 2.29.
